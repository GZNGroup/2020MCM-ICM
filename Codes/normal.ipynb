```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

snsStyle = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
snsPalette = ['deep', 'muted', 'pastel', 'bright', 'dark', 'colorblind']
sns.set(context='notebook', style=snsStyle[1], palette=snsPalette[5])
for i in snsPalette:
    sns.set_palette(i)
    sns.palplot(sns.color_palette())
```

```python
RandomSize = 1000
Normal = np.random.normal(0,1,RandomSize)
type(Normal)
```

```python
fig, axes = plt.subplots(1,2)  
sns.distplot(Normal, ax = axes[0], kde = True, rug = False)        # kde 密度曲线  rug 边际毛毯  
sns.kdeplot(Normal, ax = axes[1], shade=True)                     # shade  阴影                 
sns.despine() # 去掉多余的线
```

```python
x, y = np.random.multivariate_normal([0,0], [[1, -.5], [-.5, 1]], size=3000).T
pal = sns.dark_palette("green", as_cmap=True)
sns.kdeplot(x, y, cmap=pal)
```
